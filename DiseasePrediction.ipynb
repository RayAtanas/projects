```python
import pandas as pd
from sklearn.model_selection import train_test_split
from google.colab import drive
drive.mount('/content/drive')
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/dataset.csv')
```

```python
symptom_cols = [col for col in df.columns if col.startswith('Symptom_') and col.split('_')[-1].isdigit() and int(col.split('_')[-1]) in range(1, 18)]

# create one-hot encoded features for symptom columns
X = pd.get_dummies(df[symptom_cols])

# create a new target variable that corresponds to the 'Disease' column
y = pd.get_dummies(df['Disease'])

# split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

y_train.head()
```

```python

model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(41, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=500, batch_size=32)
```

```python
y_hat = model.predict(X_test)
y_hat = y_hat.argmax(axis=-1)
```

```python
model.save('Diseasemodel')
```
